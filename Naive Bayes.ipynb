Naive Bayes

Хотим определить спам е-мейл или нет имея дискретные фичи (кол-во такихто слов в спаме и норм письмах)

P(нормальное) = prior probability того что это норм письмо ( в нашем случае prior probability будет просто кол-во норм писем деленное на все письма) умноженное на вероятность встретить такое то значение фичи при условии что письмо нормальное (типо вероятность встретить слово друг в нормальном письме, высчитывается как кол-во слов друг в норм письмах деленное на все слова в нормальных письмах) и так по всем словам(фичам)

Потом также высчитывается вероятность того, что письмо спам и эти вероятности сравниваются 

Добавлять pseudocount к каждой фиче, ТК missing data обнулит подсчитанную вероятность целиком 


Gaussian naive bayes
Используется когда фичи continuous
Допустим есть фича рост человека, объем сердца и пытаемся предсказать понравится ему книга или нет

Строим нормальное распределение для каждой фичи (можем и другой тип распределения использовать, если видим что оно лучше описывает распределение фичи)
 
И потом уже по этим распределениям определяем вероятность ( то есть пришел человек с весом 80кг, мы смотрим на распределение - какова вероятность встретить человека с таким весом, среди тех кому понравилась книга (когда высчитывпем вероятность того, что книга понравится) и среди тех кому не понравилась (когда высчитывпем вероятность того, что книга не понравится)

Если датасет содержит и дискретные и недискретные фичи, то это непроблема, просто посчитываем каждую своим способом
