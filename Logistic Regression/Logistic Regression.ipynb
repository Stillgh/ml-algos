Model structure - linear model with sigmoid activation

Loss function - log loss

Optimization - gradient descent

Log loss - другими словами метод максимального правдоподобия используется (maximum likelihood estimation)


Например хотим предсказать съест человек апельсин или нет, фичи - кол-во съеденных апельсинов за посл год, время прошедшее с последнего приема пищи  и тп


Вот есть у нас S-shaped curve и дата поинты (люди с инфой о времени с посл приема пищи и тп)


Мы с помощью градиентного спуска (или других оптимизационных алгосов) подгоняем веса линейной функции таким образом, чтобы модель предсказывала вероятность по тому или иному человеку наиболее близко к реальной вероятности (то есть ближе к 1, если человек реально съел апельсин  и ближе к 0, если не съел, потом в формуле для тех кто не съел, из единицы вычтем предсказанную вероятность)

$y_i$ - реальная вероятность (1 или 0 - съел апельсин или нет), $p(y_i)$ - предсказанная нашей моделью
