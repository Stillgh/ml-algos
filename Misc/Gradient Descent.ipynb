Добавить формулы нахождения производной и chain rule

Оптимизационный алгоритм для нахождения минимума функции

Используется для нахождения минимума функции потерь (функции определения ошибки нашей модели, то есть насколько неверные ответы выдает наша модель, например sum of squared residuals используется как loss функция в линейной регрессии)

Соотв-но мы хотим найти такие значения параметров(лучше сказать весов) нашей модели, при которых фнукция потерь будет выдавать наименьший возможный результат

y = intercept + w1x1 + w2x2 ... wnxn, где intercept - место пересечения функцией оси  OY, w1,w2,wn - веса которые мы будем изменять, x1,x2,xn - параметры(фичи) функции. например мы хотим предсказать вес(y) имея рост(x1) и кол-во потребляемых калорий в день(x2)


SSE = (observed - y)2 (в квадрате)
SSE = (observed - (intercept + w1x1 + w2x2 ... wnxn)

Градиентный спуск будет заключаться в нахождении производных для всех параметров(intercept и веса) данной функции (то есть нахождения градиента функции)
Градиент - вектор производных
Производная - скорость изменения значения функции при изменении (стремящемся к нулю) ее параметра/параметров 

Для нахождения градиента (то есть в функции есть n параметров для которых мы хотим найти производные) используется chain rule



Chain rule также используется когда мы хотим найти производную функции не от параметра который на нее непосредственно влияет, а от параметра который влияет на параметр который влияет на значение функции

Например у нас есть данные по весу от кол-ва съеденных печенек, а так же данные по кол-ву съеденных печенек в зависимости от силы голода

Мы хотим найти производную веса от силы голода

Для этого мы находим производную веса от кол-ва съеденных печенек и производную кол-ва съеденных печенек от силы голода, а потом умножаем их и получим производную веса от силы голода
