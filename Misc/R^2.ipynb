var - variance/дисперсия вроде - среднее квадратов разницы между точкой и средним(либо мат ожиданием)
То есть сумма ря

$var = \frac{\sum_{i=1}^{n}(x-\overline{x})^2}{n-1}$

$\overline{x}$ - среднее по x

Математическое ожидание (expected value) - сумма всех возможных значений случайной величины помноженных на их вероятности
то есть X может принимать значения 1,2,3 и каждое со своей вероятностью - P(1) = 0,2; P(2) = 0,6; P(3) = 0,4

E = 1 * 0,2 + 2 * 0,6 + 3 * 0,4
