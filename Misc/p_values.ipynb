P-Value используется для определения вероятности получить такое же (или более экстремальное) значение статистики (средняя, медиана и тп) в определенном распределении случайной величины при условии того что нулевая гипотеза верна


Нулевая гипотеза - предположение о том, что между двумя явлениями не существует связи

Например мы продаем матрешек и решили проверить увеличится ли кол-во продаж, если поменяем цвет матрешек с белого на красный.
Мы собрали статистику по продажам белых матрешек.
Собрали статистику по продажам красных матрешек.

Нулевая гипотеза - смена цвета на красный не явилась причиной изменений в кол-ве продаж. 

Высчитываем p-value

Если он ниже определенного порога (обычно 0.05, alpha value), то считаем что вероятность получить такую статистику при условии верности нулевой гипотезы слишком мала и значит мы не принимаем нулевую гипотезу

Если выше порога, то возможно нулевая гипотеза верна и цвет матрешек не повлиял на кол-во продаж

Другими словами - это вероятность того что полученные значения принадлежат другому вероятностному распределению

Как считается p-value

Это сумма 3х вероятностей:
    
    1) Вероятность получить полученное значение
    2) Вероятность получить значение настолько же редкое
    3) Вероятность получить более редкое значение
    
Например бросаем монетку

Выпал 2 раза подряд орел

Нулевая гипотеза - наша монетка совершенно обычная
Альтернативная гипотеза - монетка необычная

1) Вероятность выпада орла 2 раза подряд = 0,25
2) С такой же вероятностью (0,25) может выпасть 2 решка
3) С более низкой вероятностью ничего нет (орел->решка, решка->орел у нас 0,5 вероятность (почему не понял, типо порядок не влияет на значение вероятности получить орла или решку))

Т.е. p-value = 0.25+0.25+0 = 0.5
0.5 > 0.05 значит вероятность получить такое значение (2 орла подряд) при условии верности нулевой гипотезы выше порогового значения и возможно нулевая гипотеза верна (другими словами возможно это значение не из другого распределения случайной величины, а из базового (то есть из распределения подбросов обычной (не необычной) монетки)

Короч походу если вероятность наблюдаемого события/полученной величины слишком мала, то мы говорим что значение получено из другого распределения/наше изменение (покрасили матрешку) повлияло на величину
