Регуляризация используется для борьбы с переобучением и заключается в уменьшении значимости фичей. Она увеличит bias (ухудшит перфоманс модели на трейнинг сете), но уменьшит variance (улучшит перфоманс на новых данных)

L2/Squared/Ridge regression - 

добавляет переменную пенальти - $\lambda$ к функции затрат (в случае линейной регрессии мы пытаемся уменьшить SSR (sum of squared residuals), при добавление L2 регрессии мы будем пытаться уменьшить SSR + $\lambda$ * $slope^2$ (где slope - сумма квадратов всех фичей умноженных на веса, типо ($w_1*рост)^2 + (w_2*вес)^2$ и тд). Intercept не включается, тк он напрямую не влияет на то как какая-либо фича влияет на y

Чем меньше фича влияет на таргет, тем сильнее регуляризация ее уменьшит и чем больше фича влияет, тем меньше она уменьшится 

Если у нас только одна фича используется, то оптимальное значение для пенальти можно подобрать градиентным спуском, если больше, то более сложные методы используются

L1/Absolute/Lasso regression -

Она использует не квадрат как в ridge, а абсолютное значение

Разница между Lasso и Ridge состоит в том, что Ridge будет приближать значение фичей к 0, но не обнулит их, а Lasso может обнулить.

Lasso лучше использовать когда больше неважных фич, Ridge когда больше важных

Elastic net regression - 

Когда фич дофига и мы не знаем какие из них полезны, а какие нет, то юзаем elastic net 

SSR + $\lambda_1 * |slope| + \lambda_2 * slope^2 $

Она объединяет L1 и L2 регрессии + она лучше работает с correlated параметрами (потому что l1 обычно берет один из скорреллированных параметров, а второй обнуляет, а l2 уменьшает значения всех скорреллированных параметров)
